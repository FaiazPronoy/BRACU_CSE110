```python
#T7
def showPalindrome(num):
    line = ''
    for idx in range(1,num+1):
       line += str(idx)
    for idx2 in range(num-1,0,-1):
        line += str(idx2)
    return line

num=int(input('Enter the amount: '))
print(showPalindrome(num))
```

```python
#T8
def showPalindrome(num, separator = ""):
    line = ""
    for idx in range(1,num+1):
       line += str(idx) +separator
    for idx2 in range(num-1,0,-1):
        line += str(idx2) +separator
    return line

def show_palindromic_triangle(num):
  for iter in range(1,num+1):
    print(" "*2*(num-iter) + showPalindrome(iter, " "))
num=int(input('Enter the amount: '))
show_palindromic_triangle(num)
```

```python
#T13
def numfuction(opt,num1,num2):
    if opt =='+':
        res=num1+num2
        return res
    elif opt=='-':
        res=num1-num2
        return res
    elif opt=='*':
        res=num1*num2
        return res
    elif opt=='/':
        res=num1/num2
        return res
opt=input('Enter oparetor: ')
num1=float(input('Enter 1st number: '))
num2=float(input('Enter 2nd number: '))

print(numfuction(opt,num1,num2))
```

```python
#T14
def cuttingcharacters(sentence,position):
    empty=''
    counter=''
    for idx in range(0,len(sentence)):
        if idx!=0 and idx%position==0:
            empty+=sentence[idx]
        else:
            counter+=sentence[idx]

    return counter+empty
sentence=str(input('Enter sentence: '))
position=int(input('Enter position: '))
print(cuttingcharacters(sentence,position))
```

```python
#T18
def division(starting,ending,first_divisor,second_divisor):
    sum=0
    for i in range(starting,ending):
        if i%first_divisor==0 and i%second_divisor==0:
            continue
        elif i%first_divisor==0:
            sum=sum+i
        elif i % second_divisor == 0:
            sum = sum + i
    return sum
starting=int(input())
ending=int(input())
first_divisor=int(input())
second_divisor=int(input())

print(division(starting,ending,first_divisor,second_divisor))
```
