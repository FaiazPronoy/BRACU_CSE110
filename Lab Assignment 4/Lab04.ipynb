```python
#Task_04
inp = input('Enter a list: ')
inp = inp[1:-1]
list1= inp.split(', ')
list2= []
for elem in list1 :
  list2.append(int(elem))
for count in range(len(list2)) :
  list2[count]=list2[count]**2
print(list2)
```

```python
#Task_06
inp = (input("Enter a list: "))
inp = inp[1:-1]
list1 = inp.split(', ')
list2=[]
for elem in list1 :
  list2.append(int(elem))
maximum= None
for count in range(len(list2)):
  if maximum is None or list2[count]>maximum :
    maximum = list2[count]
print('Largest number is', maximum, 'Which was found at index', list2.index(maximum))
```
