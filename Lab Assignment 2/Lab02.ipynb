```python
#Task_12

n = int(input())
x = n
track = 0
while n != 0:
  n = n // 10
  track += 1

power = 10**(track - 1)

while power > 0 :
  result = x//power
  x=x%power
  power = power // 10

  if power == 0 :
    print(result)
  else :
    print(result, end= ", ")
```
