```python
#Task_03
s = input()

for ch in s:
    if ch != '1' and ch != '0':
        print('Not a Binary Number')
        break
else:
    print('Binary Number')
```

```python
#Task_03(Another Way)
sent=input()
not_binary = False

for char in sent :
  if char != '0' and char != '1' :
    not_binary= True
    break
if not_binary == True :
  print("Not a binary number")
else: 
  print("Binary number")
```

```python
#Task_07
string = str(input("enter: "))

for num in string:
    acsii = ord(num)

    if acsii == 122:
        next_alpha = acsii - 25
    else:
        next_alpha = acsii + 1




    final = chr(next_alpha)
    print(final, end='')
```

```python
#Task_08
src = input()
res = ''

i = 0
shft = ord('a') - ord('A')
for ch in src:
    if i%2 == 1:
        ch_up = chr(ord(ch) - shft)
        res += ch_up
    i += 1

print(res)
```

```python
#Task_10
s1 = input("enter: ")
s2 = input("enter: ")

le1 = len(s1)
le2 = len(s2)

if le1>le2:
    s3 = s1[0:le2]
    for i in range(le2):
        print(s3[i],end='')
        print(s2[i],end='')
    print(s1[le2:],end='')
if le2>le1:
    s3 = s2[0:le1]
    for i in range(le1):
        print(s1[i],end='')
        print(s3[i],end='')
    print(s2[le1:],end='')
if le1 == le2:
    for i in range(le1):
        print(s1[i],end='')
        print(s2[i],end='')
```
