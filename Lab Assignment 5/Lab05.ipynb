```python
#Task_08
dict1 = {}
str1 = "{'Jon': 100, 'Dan': 200, 'Rob': 30, 'Ned': 110}"
str1 = str1[1:-1]

list1 = str1.split(", ")

for item in list1:
  list2 = item.split(": ")
  key = list2[0][1:-1]
  val = int(list2[1])
  dict1[key] = val
print(dict1)

total = 0
counter = 0
for v in dict1.values(): 
  total += v
  counter += 1
  average = int(total/counter)
print("Average is",average)
```

```python
#Task_14
dict1 = {'a':59 , 'b':-82 , 'c':5 , 'd':-81 , 'e':53}
for i in dict1:
    j = 0
    k = 22
    while j < 5:
        if j % 2 == 0:
            k = dict1[i] + j - (8 + k % 6) / 3
            dict1[i] = dict1[i]+ int(k)
        else:
            k = dict1[i] + j - (6 - k % 8) * 3
            dict1[i] = dict1[i] - int(k)
        j += 1
    print(int(k))
    print(i + " -> " + str(dict1[i]))
```

```python
#Task_20
user1 = input("")
user1 = user1.replace('.',"")
user1 = user1.split(" ")
user2 = ['@', '$', '&', '#']
my_dict = {}
for index in user1:
  x = 0 
  for idx in index:
    x+= ord(idx)
  i = x % len(user2) 
  if user2[i] not in my_dict:
    my_dict[user2[i]]=[index]
  elif user2[i] in my_dict and index not in my_dict[user2[i]]:
    my_dict[user2[i]].append(index)

print("Words in the given String:", list(user1))

print("Answer: ", my_dict)
```
