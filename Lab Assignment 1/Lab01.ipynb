```python
#Task_13
mark = int(input("Please enter a student's mark to know the corresponding grade= "))

if 90 <= mark <= 100 :
  print("Grade= A")

elif 80 <= mark < 90 :
  print("Grade= B")

elif 70 <= mark < 80 :
  print("Grade= C")

elif 60 <= mark < 70 :
  print("Grade= D")

elif 50 <= mark < 60 :
  print("Grade= E")

elif 0 <= mark < 50 :
  print("Grade= F")

else:
  print("Invalid Mark")
```

```python
#Task_14
distance = int(input("Please enter the distance in meters= "))
time = int(input("Please enter the time in seconds= "))

if 0 < (distance / 1000) / (time / 3600) < 60 :
  print ((distance / 1000) / (time / 3600),"km/h")
  print("Too slow.Needs more changes.")

elif 60 <= (distance / 1000) / (time / 3600) <= 90 :
  print ((distance / 1000) / (time / 3600),"km/h")
  print("Velocity is okay. The car is ready!")

elif 90 < (distance / 1000) / (time / 3600) :
  print ((distance / 1000) / (time / 3600),"km/h")
  print("Too fast. Only a few changes should suffice.")

else:
  print("Invalid")
```
